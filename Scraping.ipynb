## Import Library dan Package yang dibutuhkan

```python
import googleapiclient.discovery
import pandas as pd

from googleapiclient.discovery import build
from google.colab import files
```

## Scraping Data

```python
# inisialisasi platform
service_name = 'youtube'
service_version = 'v3'
api_key_dev = api_key

yt_api = build(service_name, service_version, developerKey=api_key_dev)
```

```python
video_id = '1qDL8bPuj_I'

# List untuk menyimpan komentar
all_comments = []
next_page_token = None

# Looping untuk ambil semua halaman komentar
while True:
    request = yt_api.commentThreads().list(
        part='snippet',
        videoId=video_id,
        maxResults=100,
        pageToken=next_page_token  # token untuk halaman selanjutnya
    )

    # Jalankan request
    response = request.execute()

    # Loop untuk ambil data dari setiap komentar di halaman ini
    for entry in response['items']:
        snippet = entry['snippet']['topLevelComment']['snippet']
        is_public = entry['snippet']['isPublic']
        all_comments.append([
            snippet['authorDisplayName'],
            snippet['publishedAt'],
            snippet['likeCount'],
            snippet['textDisplay'],
            is_public
        ])

    # Cek apakah ada halaman selanjutnya
    next_page_token = response.get('nextPageToken')
    if not next_page_token:
        break
```

## Menyimpan Data ke dalam Dataframe

```python
# membuat dataframe
comments_df = pd.DataFrame(all_comments, columns=['author', 'published_time', 'likes', 'comment', 'isPublic'])
comments_df.info()
```
